# Чтение и запись файлов

```python
import os
tempDir = os.path.abspath('Temp')
if not os.path.exists(tempDir):
    os.mkdir(tempDir)
tempFile = os.path.join(tempDir, 'Temp.txt')
print(tempFile)
if not os.path.exists(tempFile):
    open(tempFile, 'w').write('Test string')
```

```python
f = open(tempFile)
for line in f:
    print(line)
f.close()

print('=' * 100 )

with open(tempFile) as f:
    print(f.readline())
    print(f.read())
```
