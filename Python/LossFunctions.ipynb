```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
from ipywidgets import *

# Чтобы графики были побольше
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
x = np.linspace(-3, 4, 200)

# синяя
squareLoss = np.square(1 - x)
plt.plot(x, squareLoss)

# зелёная
indicatorLoss = (1 - np.sign(x)) / 2
plt.plot(x, indicatorLoss)

# красная
logisticLoss = np.log(1 + np.exp(-x)) # / math.log(2)
plt.plot(x, logisticLoss)

# бирюзовая
hingeLoss = np.maximum(0, 1 - x)
plt.plot(x, hingeLoss)

# фиолетовая
adaboostLoss = np.exp(-x)
plt.plot(x, adaboostLoss)

plt.xlim(x[0], x[-1])
plt.ylim(-1, 4)
plt.axes().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
squareLossD = np.gradient(squareLoss, x)
plt.plot(x, squareLossD)

indicatorLossD = np.gradient(indicatorLoss, x)
plt.plot(x, indicatorLossD)

logisticLossD = np.gradient(logisticLoss, x)
plt.plot(x, logisticLossD)

hingeLossD = np.gradient(hingeLoss, x)
plt.plot(x, hingeLossD)

adaboostLossD = np.gradient(adaboostLoss, x)
plt.plot(x, adaboostLossD)

plt.xlim(x[0], x[-1])
plt.ylim(-3, 2)
plt.axes().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
squareLossDD = np.gradient(squareLossD, x)
plt.plot(x, squareLossDD)

indicatorLossDD = np.gradient(indicatorLossD, x)
plt.plot(x, indicatorLossDD)

logisticLossDD = np.gradient(logisticLossD, x)
plt.plot(x, logisticLossDD)

hingeLossDD = np.gradient(hingeLossD, x)
plt.plot(x, hingeLossDD)

adaboostLossDD = np.gradient(adaboostLossD, x)
plt.plot(x, adaboostLossDD)

plt.xlim(x[0], x[-1])
plt.ylim(-1, 4)
plt.axes().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
def drawLoss(y):
    x = np.linspace(-3, 4)
    logLoss = y * np.log(1 + np.exp(-x)) + (1 - y) * np.log(1 + np.exp(x))
    plt.plot(x, logLoss)
    plt.xlim(x[0], x[-1])
    plt.ylim(-1, 4)
    plt.axes().set_aspect('equal')
    plt.grid(True)
    plt.show()

interact(drawLoss, y=widgets.FloatSlider(min=-2, max=2, step=0.1, value=0))
```
